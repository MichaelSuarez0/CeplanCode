```python
import requests

# URL base (excluyendo el parámetro offset para actualizarlo con cada página
base_url = "https://sires.unsch.edu.pe/api/emails/20?search=&offset={}&limit=100"

# List to hold all the rows from JSON responses
all_rows = []

# Define the range for the offset (incrementos de 100, hasta 1200)
for offset in range(0, 1300, 100):
    # Construct the full URL by inserting the current offset
    api_url = base_url.format(offset)
    
    # Send a GET request to the API
    response = requests.get(api_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response into a dictionary
        data = response.json()
        # Extract the rows and append to the list
        rows = data.get('rows', [])
        print(rows)
        all_rows.extend(rows)
    else:
        print(f"Failed to retrieve data for offset {offset}. Status code: {response.status_code}")

```

```python
import pandas as pd
#print(all_rows)

# Solo falta limpiar la data
directorio = pd.DataFrame(all_rows)
directorio = directorio.rename(columns ={
    "datos": "Apellidos y Nombres",
    "celular": "Celular",
    "email": "Email"
})

#print(directorio)

#directorio = directorio.drop(columns = ["Celular"])
directorio.to_excel("Directorio_UNSCH.xlsx", sheet_name="Hoja 1", index=False)
```

```python
def añadir_apelativo(lista_nombres, masculinos, femeninos):
    apelativo = []

    for nombre_completo in lista_nombres:
        # Limpiar y dividir el nombre en partes individuales
        partes_nombre = nombre_completo.strip().split()
        #print(partes_nombre)
        
        # Variable para determinar el apelativo
        apelativo_nombre = "Sr./Sra."
        
        for parte in reversed(partes_nombre):
            # Verificar si la parte del nombre está en las listas de masculinos o femeninos
            if parte in masculinos:
                apelativo_nombre = "Sr."
                break  # No es necesario seguir buscando si ya se encontró un masculino
            elif parte in femeninos:
                apelativo_nombre = "Sra."
                break  # No es necesario seguir buscando si ya se encontró una femenina
            
        
        apelativo.append(apelativo_nombre)
    
    # Crear un DataFrame con los nombres y apelativos
    df = pd.DataFrame({
        'Nombre': lista_nombres,
        'Apelativo': apelativo
    })
    
    return df

# Suponiendo que tienes una lista de nombres y tienes nombres_h_clean y nombres_m_clean definidos previamente
# Lista de nombres de ejemplo
lista_nombres = df["Apellidos y Nombres"].str.lower()
nombres_masculinos = df2[df2['genero'] == 'masculino']['nombre'].tolist()
nombres_femeninos = df2[df2['genero'] == 'femenino']['nombre'].tolist()
df_apelativos = añadir_apelativo(lista_nombres, nombres_masculinos, nombres_femeninos)

print(len(df_apelativos))
print(df_apelativos)
```
